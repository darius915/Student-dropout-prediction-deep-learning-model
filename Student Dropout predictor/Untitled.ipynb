```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('dataset.csv')
```

```python
data.dropna(inplace =True)
data
```

```python
#data.drop(columns =['GDP','Inflation rate','Unemployment rate'],inplace =True)
```

```python
data
```

```python
data['Target'].value_counts()
```

```python
data.drop(data.index[data['Target'] == 'Enrolled'], inplace = True)
```

```python
data['Target'].value_counts()
```

```python
new = {'Dropout':0,
      'Graduate':1}
data['Target'] =data['Target'].replace(new)
```

```python
data
```

```python
train_data = data.sample(frac = 0.8, random_state=0)
test_data = data.drop(train_data.index)
test_data.shape
```

```python
train_label = train_data.pop('Target')
test_label = test_data.pop('Target')
test_label.shape
```

```python
x_train = tf.keras.utils.normalize(train_data)
y_test = tf.keras.utils.normalize(test_data)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(800,activation =tf.nn.relu))
model.add(tf.keras.layers.Dense(400,activation =tf.nn.relu))
model.add(tf.keras.layers.Dense(200,activation =tf.nn.relu))
model.add(tf.keras.layers.Dense(100,activation =tf.nn.relu))
model.add(tf.keras.layers.Dense(1))
```

```python
model.compile(optimizer ='adam',
             loss = 'mse',
             metrics =['mae','mse'])
```

```python
model.fit(x_train,train_label,epochs =100)
```

```python
loss,mae,mse = model.evaluate(y_test,test_label,verbose=0)
print('Testing set Mean Abs Error :{:5.2f}'.format(mae))
```

```python
model.save('school_dropout.model')
```
